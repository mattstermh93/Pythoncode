### Shopping cart program
One is without multiple functions, the other is with functions to minimize repeat code

```python
#Without multiple functions defined and called

from IPython.display import clear_output
    
def displayInstructions():
    print('You can type add to add items to your shopping cart')
    print('You can type delete to delete items from your shopping cart')
    print('You can type show to see what is currently in your shopping cart')
    print('You can type quit to exit the program and see what is in your shopping cart')
    

def shoppingCart():
    food = ['grapes', 'olives', 'donuts', 'bread', 'cheese', 'milk', 'salad', 'burgers', 'tomatos', 'potatos', 'almonds', 'eggs',
   'olive oil', 'peaches', 'plums', 'garlic', 'pepper', 'salt', 'bananas', 'chicken', 'meat', 'soy', 'macaroni', 'pasta',
   'sauce', 'toilet paper', 'tissues', 'orange juice', 'grape juice', 'wine', 'beer', 'liquor', 'tequila', 'grapefruit', 'mango']
    
    shopping_list = []
    while True:
        what_item = input('What would you like to add? : ')
        if what_item not in food:
            print('Sorry, I do not understand. Please try again. ')
            continue
        shopping_list += [what_item]     
        user_input = input('What else would you like to do? ')
        if user_input == 'add':
            continue
        elif user_input == 'quit':
            print('Have a great day! Here are your items {} '.format(shopping_list))
            break
        elif user_input == 'show':
            print('Here are your items\n \n ---------- \n ---------- \n  \n {} '.format(shopping_list))
            anything_more = input('Is there anything else you would like to do? Type add to add more items: ')
            if anything_more == 'no':
                print('Ok, have a great day! Here are the items in your cart: {}'.format(shopping_list))
                break
            elif anything_more == 'quit':
                print('Have a great day! Here are your items {} '.format(shopping_list))
                break
        elif user_input == 'delete':
            delete_what = input('what would you like to delete? ')
            if delete_what not in shopping_list:
                print('That item is not in the list!')
                continue
            shopping_list.remove(delete_what)
            what_else = input("Is there anything else you would like to do? If not, feel free to type quit")
            if what_else == 'quit':
                print('Have a great day! Here are your items {} '.format(shopping_list))
                break
            elif what_else == 'show':
                print('Here are your items\n \n ---------- \n ---------- \n  \n {} '.format(shopping_list))
                anything_more = input('Is there anything else you would like to do? Type add to add more items: ')
                if anything_more == 'no':
                    print('Ok, have a great day! Here are the items in your cart: {}'.format(shopping_list))
                    break
                else:
                    print('I don\'t understand - please try again!')
                    continue



shoppingCart()
```

```python
# With newer functions

from IPython.display import clear_output
    
def displayInstructions(shopping_list):
    print('Type add to add an item to your cart')
    print('You can type delete to delete items from your shopping cart')
    print('You can type show to see what is currently in your shopping cart')
    print('You can type quit to exit the program and see what is in your shopping cart')
    
def addItem(shopping_list, food):
    what_item = input('What would you like to add? : ')
    shopping_list.append(what_item)
    print('{} has been added'.format(what_item))
    if what_item not in food:
        print('Sorry, I do not understand. Please try again. ')
          
def showItems(shopping_list):
    print('Here are your items\n \n ---------- \n ---------- \n  \n {} '.format(shopping_list))
    
def deleteItems(shopping_list):
    delete_what = input('what would you like to delete? ')
    if delete_what not in shopping_list:
        print('That item is not in the list!')
    else:
        shopping_list.remove(delete_what)
                
def quitProgram(shopping_list):
    want_to_quit = input("Do you want to quit? Type yes or no")
    print('Have a great day! Here are your items \n----------\n---------- \n {} '.format(shopping_list))
    if want_to_quit == 'yes':
        return True
    elif want_to_quit == 'no':
        return False
    

    

def shoppingCart():
    
    
    food = ['grapes', 'olives', 'apples', 'oranges', 'donuts', 'bread', 'cheese', 'milk', 'salad', 'burgers', 'tomatos', 'potatos', 'almonds', 'eggs',
   'olive oil', 'peaches', 'plums', 'garlic', 'pepper', 'salt', 'bananas', 'chicken', 'meat', 'soy', 'macaroni', 'pasta',
   'sauce', 'toilet paper', 'tissues', 'orange juice', 'grape juice', 'wine', 'beer', 'liquor', 'tequila', 'grapefruit', 'mango']
    
    
    shopping_list = []
    displayInstructions(shopping_list)
    while True:
        
        what_to_do = input ("What would you like to to do? ")
        if what_to_do == 'add':
            addItem(shopping_list, food)
            continue
        elif what_to_do == 'quit':
            if quitProgram(shopping_list) == True:
                break
            else:
                continue
        elif what_to_do == 'show':
            showItems(shopping_list)
        elif what_to_do == 'delete':
            deleteItems(shopping_list)
            continue                



shoppingCart()
```
